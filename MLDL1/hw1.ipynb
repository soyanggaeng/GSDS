### 1.(a) Fit a multiple linear regression model to predict Sales using Price, Urban, and US. Report the R^2 of the model.

```python
import pandas as pd
carseats = pd.read_csv("Carseats.csv")
carseats
```

```python
carseats.describe()
```

```python
print(carseats['Urban'].value_counts())
print(carseats['US'].value_counts())
```

Urban과 US의 값이 qualitative하므로 이를 Yes=1, No=0으로 바꿔줌

```python
carseats.replace({'Urban':'Yes'},1, inplace=True)
carseats.replace({'Urban':'No'},0, inplace=True)
carseats.replace({'US':'Yes'},1, inplace=True)
carseats.replace({'US':'No'},0, inplace=True)
print(carseats['Urban'].value_counts())
print(carseats['US'].value_counts())
```

```python
import statsmodels.formula.api as sm
cs_model = sm.ols(formula='Sales ~ Price + Urban + US', data=carseats).fit()
cs_model.summary()
```

참고: https://www.reneshbedre.com/blog/linear-regression.html

**같은 작업을 skikit learn으로 할 경우**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
X = carseats[['Price', 'Urban', 'US']]
Y = carseats['Sales']
sk_model = LinearRegression().fit(X, Y)
print("r2_score:", r2_score(Y, sk_model.predict(X)))
print("coefficients of Price, Urban, US:", sk_model.coef_)
print("intercept:", sk_model.intercept_)
```

#### ① Split the sample set into a training set and a validation set.

```python
from sklearn.model_selection import train_test_split
import numpy as np
import time
X = default[['income', 'balance']]
y = default['default']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.5, random_state=np.random.seed(int(time.time())))
```

#### ② Fit a multiple logistic regression model using only the training observations.
참고: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression.predict_proba

```python
from sklearn.linear_model import LogisticRegression
mlr = LogisticRegression(random_state=np.random.seed(int(time.time()))).fit(X_train, y_train)
```

#### ③ Obtain a prediction of default status for each individual in the validation set by computing the posterior probability of default for that individual, and classifying the individual to the default category if the posterior probability is greater than 0.5.

```python
mlr.predict_proba(X_val)
```

```python
val_set = pd.concat([X_val, y_val], axis=1)
val_set.head()
```

```python
df_val = pd.DataFrame(mlr.predict_proba(X_val), index=val_set.index, columns=['post_prob_0', 'post_prob_1'])
df_val.head()
```

```python
post_prob = pd.concat([val_set, df_val], axis=1)
post_prob.head()
```

```python
post_prob.reset_index(inplace=True, drop=True)
post_prob['prediction'] = -1
for i in range(len(post_prob)):
    if post_prob.iloc[i]['post_prob_1'] <= 0.5:
        post_prob.loc[i, 'prediction'] = 0
    elif  post_prob.iloc[i]['post_prob_1'] > 0.5:
        post_prob.loc[i, 'prediction'] = 1
post_prob.head()
```

```python
print(post_prob['default'].value_counts())
print(post_prob['prediction'].value_counts())
```

### 2.(c) Perform 5-fold cross-validation using the same model as in Part (a). Write your own code for the K-fold cross-validation. How does the validation error of the cross-validation the differ from to the results in Part (a)?

```python
import pandas as pd
from typing import List
def Kfold(df:pd.DataFrame, X_name:List, Y_name:str, k:int):
    # shuffle
    import numpy as np
    import time
    df_shuffled = df.iloc[np.random.RandomState(seed=int(time.time())).permutation(df.index)]
    v_error = []
    fp_rate = []
    fn_rate = []
    AUC = []
    
    X = df_shuffled[X_name]
    y = df_shuffled[Y_name]
    
    for i in range(k):
        # split train & validation
        X_train = X.drop(df_shuffled.index[int(len(df_shuffled)/k)*i:int(len(df_shuffled)/k)*(i+1)])
        X_val = X[int(len(df_shuffled)/k)*i:int(len(df_shuffled)/k)*(i+1)]
        y_train = y.drop(df_shuffled.index[int(len(df_shuffled)/k)*i:int(len(df_shuffled)/k)*(i+1)])
        y_val = y[int(len(df_shuffled)/k)*i:int(len(df_shuffled)/k)*(i+1)]        

        print("[",i+1,"iteration]\n",'- X_train(row:',len(X_train),')\n',X_train.head(),'\n- X_val(row:',len(X_val),'):\n',
              X_val.head(),'\n')
        
        # fit model
        from sklearn.linear_model import LogisticRegression
        kfold_model = LogisticRegression(random_state=np.random.seed(int(time.time()))).fit(X_train,y_train)
        
        # do the step 3 in 2.(a)
        val_set = pd.concat([X_val, y_val], axis=1)
        df_val = pd.DataFrame(kfold_model.predict_proba(X_val), index=val_set.index, columns=['post_prob_0', 'post_prob_1'])
        post_prob = pd.concat([val_set, df_val], axis=1)
        post_prob.reset_index(inplace=True, drop=True)
        post_prob['prediction'] = -1
        for i in range(len(post_prob)):
            if post_prob.iloc[i]['post_prob_1'] <= 0.5:
                post_prob.loc[i, 'prediction'] = 0
            elif  post_prob.iloc[i]['post_prob_1'] > 0.5:
                post_prob.loc[i, 'prediction'] = 1
                
        print("label counts:\n", post_prob['default'].value_counts(), "\n",
              post_prob['prediction'].value_counts(), "\n"+'+'*20+'\n')
        
        # Compute the validation set error
        from sklearn.metrics import accuracy_score
        v_error.append(1-accuracy_score(post_prob['default'], post_prob['prediction']))
        
        # Compute false positive rate
        from sklearn.metrics import confusion_matrix
        cf = pd.DataFrame(confusion_matrix(post_prob['default'], post_prob['prediction'], labels=[0, 1]), 
                     index=['No(true=0)', 'Yes(true=1)'], columns=['No(predicted=0)', 'Yes(predicted=1)'])
        fp_rate.append(cf['Yes(predicted=1)']['No(true=0)']/(cf['No(predicted=0)']['No(true=0)']+cf['Yes(predicted=1)']['No(true=0)']))
        #Compute false negative rate
        fn_rate.append(cf['No(predicted=0)']['Yes(true=1)']/(cf['No(predicted=0)']['Yes(true=1)']+cf['Yes(predicted=1)']['Yes(true=1)']))
        
        #Compute AUC
        from sklearn.metrics import roc_auc_score
        AUC.append(roc_auc_score(post_prob['default'], post_prob['prediction']))
        
    cvk, fp, fn, auc = 0, 0, 0, 0
    for j in range(len(v_error)):
        nk = int(len(df_shuffled)/k)
        n = len(df_shuffled)
        print("for model",j+1,": \n\tvalidation error is", v_error[j], "\n\tfalse positive rate is", fp_rate[j], 
             "\n\tfalse negative rate is", fn_rate[j], "\n\tAUC is", AUC[j])
        cvk += (nk*v_error[j])/n
        fp += (nk*fp_rate[j])/n
        fn += (nk*fn_rate[j])/n
        auc += (nk*AUC[j])/n
    print("\n\nvalidation error is:",cvk,"\nfalse positive rate is:",fp,"\nfalse negative rate is:",fn,
         "\nAUC is:",auc)
```

```python
Kfold(default, ['income', 'balance'], 'default', 5)
```
